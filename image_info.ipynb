```python
import pandas as pd
import glob
from PIL import Image
from collections import Counter
```

```python
train_data_dir = '../Train_label.csv'
test_data_dir = '../submit_example.csv'

train_df = pd.read_csv(train_data_dir)
train_img_paths = list(train_df['FileName'])
local_path = '../Train/'
for i in range(len(train_img_paths)):
    train_img_paths[i] = local_path + train_img_paths[i]
    
test_df = pd.read_csv(test_data_dir)
test_img_paths = list(test_df['FileName'])
local_path = '../Test/'
for i in range(len(test_img_paths)):
    test_img_paths[i] = local_path + test_img_paths[i]
```

```python
img_paths = train_img_paths + test_img_paths
```

```python
size_1 = []
size_2 = []
for i in img_paths:
    img = Image.open(i)
    size_1.append(img.size[0])
    size_2.append(img.size[1])
```

```python
bili = []
for c,k in zip(size_1,size_2):
    b =c/k
    bili.append(b)
bili_0 = [round(i,0) for i in bili]
bili_0
```

```python
bili_0= Counter(bili_0)
bili_0
```

```python
bili_1 = [round(i,1) for i in bili]
bili_1= Counter(bili_1).most_common(10)
bili_1
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.bar(range(len(bili)), bili)
plt.show()
```
