```python
import os
os.environ["CUDA_VISIBLE_DEVICES"] = "-1"

from warmup_cosine_decay_scheduler import WarmUpCosineDecayScheduler
learning_rate = 1e-3
sample_count =5760
epochs = 50
warmup_epoch = 10
batch_size = 32
learning_rate_base = learning_rate
total_steps = int(epochs * sample_count / batch_size)
warmup_steps = int(warmup_epoch * sample_count / batch_size)

warm_up_lr = WarmUpCosineDecayScheduler(learning_rate_base=learning_rate_base,
                                        total_steps=total_steps,
                                        warmup_learning_rate=0,
                                        warmup_steps=warmup_steps,
                                        hold_base_rate_steps=0,
                                        )
```

```python
from keras.models import Sequential
from keras.layers import Dense
# Create a model.
model = Sequential()
# model.add(Dense(32, activation='relu', input_dim=100))
model.add(Dense(10, activation='softmax'))
model.compile(optimizer='rmsprop',
            loss='categorical_crossentropy',
            metrics=['accuracy'])
```

```python
import numpy as np
from tensorflow import keras
from keras import backend as K
data = np.random.random((sample_count, 100))
labels = np.random.randint(10, size=(sample_count, 1))

one_hot_labels = keras.utils.to_categorical(labels, num_classes=10)
```

```python
from keras.callbacks import TensorBoard, Callback, ModelCheckpoint

log_local = './logs/'+'L0002_test_lr'
tensorBoard = TensorBoard(log_dir=log_local)
model.fit(data, one_hot_labels, epochs=100, batch_size=batch_size,
        verbose=1, callbacks=[warm_up_lr,tensorBoard])
```
